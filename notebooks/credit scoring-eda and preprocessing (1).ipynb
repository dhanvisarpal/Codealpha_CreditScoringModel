```python
# Load CSV file
df = pd.read_csv("default of credit card clients.csv", header=1)

#print first 5 rows 
print(df.head())

#print last 5 rows 
print("\n", df.tail())
```

```python
#target variable distribution 
import seaborn as sns
import matplotlib.pyplot as plt
#class distribution
print(df["default payment next month"].value_counts())
#percent of class
print(df["default payment next month"].value_counts(normalize=True) * 100)
plt.figure(figsize=(8,6))

sns.countplot(
    x="default payment next month",
    data=df
)

plt.title("Default Payment Distribution")
plt.xlabel("Default Payment")
plt.ylabel("Count")

plt.show()
```

```python
#heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df.corr())
plt.title("Correlation Heatmap")
plt.show()
```

```python
#dropping id column 
if "ID" in df.columns:
    df.drop("ID", axis=1, inplace=True)
    print("\nID column removed.")
```

```python
#histogram
df.hist(figsize=(18,18))
plt.tight_layout()
plt.show()
```

```python
df.skew(numeric_only=True).sort_values(ascending=False)
```

```python
cols = [
    "BILL_AMT1","BILL_AMT2","BILL_AMT3","BILL_AMT4","BILL_AMT5","BILL_AMT6",
    "PAY_AMT1","PAY_AMT2","PAY_AMT3","PAY_AMT4","PAY_AMT5","PAY_AMT6"
]
#negative values
(df[cols] < 0).sum()
```

```python
for col in cols:
    if (df[col] < 0).sum() == 0:
        df[col] = np.log1p(df[col])
        print(f"{col}: Log transformation applied")
    else:
        print(f"{col}: Contains negative values, skipped")
```

```python
#outlier detection
numerical_cols = [
    'LIMIT_BAL', 'AGE',
    'PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6',
    'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3',
    'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6',
    'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3',
    'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6'
]
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = ((df[col] < lower) | (df[col] > upper)).sum()

    print(f"{col}: {outliers} outliers")
```

```python
plt.figure(figsize=(18,10))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(5, 4, i)
    sns.boxplot(y=df[col])
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
print("\nCategorical Value Counts\n")

categorical_cols = ["SEX", "EDUCATION", "MARRIAGE"]

for col in categorical_cols:
    print(f"\n{col}")
    display(df[col].value_counts().sort_index())
```

```python
# Correct invalid education and marriage values
df["EDUCATION"] = df["EDUCATION"].replace({0: 4, 5: 4, 6: 4})
df["MARRIAGE"] = df["MARRIAGE"].replace({0: 3})
print("Invalid categorical values corrected.")
```

```python
df.to_csv("processed_credit_card.csv", index=False)
print("saved successfully")
```
